```python
# Definimos la funcion euler 
def euler (y,t,dt, derivs):
    y_next = y + derivs(y,t)*dt
    return y_next
```

```python
N = 1000  #Numero de pasos    
xo = 0.0  #Posicion inicial 
vo = 0.0  #Velocidad Inicial 
tau = 3.0 #Tiempo total en segundos  

dt = tau / float (N - 1) #paso en tiempo 
k = 3.5  #Constante del resorte en [N/m]
m = 0.2 #masa en kilogramos 
gravity = 9.8 
time = linspace( 0 , tau , N) #Tiempo 

# 
```

Cree una matriz Nx2 para almacenar los resultados de nuestros cálculos. Cada fila de 2 elementos se utilizará para el estado del sistema en un instante, y cada instante estará separado por tiempo. El primer elemento en cada fila será la posición, el segundo la velocidad.

```python
y = zeros([N,2]) 
y[0,0] = xo
y[0,1] = vo

def SHO( state , time ) :
    g0 = state[1]
    g1 = -k/m * state [0] - gravity
    return array ([g0,g1])

for j in range (N-1):
    y[j+1] = euler(y[j], time[j], dt, SHO)
    
xdata = [ y [ j , 0 ] for j in range (N ) ]
vdata = [ y [ j , 1 ] for j in range (N ) ]

plot( time , xdata, c = "red" )
plot( time , vdata, c = "purple")
xlabel( "time" )
plt.grid()
ylabel ("position , velocity")
show ( )


```

# Example 4.4.1

Metodo Runge - Kutta segundo orden 
    

Primeros 3 graficas con K cercana a 0

```python
def rk2( y , time , dt , derivs):
    k0 = dt * derivs( y , time )
    k1 = dt * derivs( y + k0 , time + dt )
    y_next = y + 0.5 * (k0+k1)
    return y_next

k = .1

for j in range (N-1):
    y[j+1] = rk2(y[j], time[j], dt, SHO)
    
xdata = [ y [ j , 0 ] for j in range (N ) ]
vdata = [ y [ j , 1 ] for j in range (N ) ]

plot( time , xdata, c = "red" )
plot( time , vdata, c = "purple")
xlabel( "time" )
plt.title("K = .1")
plt.grid()
ylabel ("position , velocity")
show ( )
```
